# Task 2 - Aggregation of Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Read CSV files into pandas DataFrames
trips_count_df = pd.read_csv("./trips_count/trips_count.csv")
platform_profits_df = pd.read_csv("./platform_profits/platform_profits.csv")
driver_earnings_df = pd.read_csv("./driver_earnings/driver_earnings.csv")
```

### 2.1 Counting Number of Trips for each Month

```python
# sort values by month
trips_count_df.sort_values(by='month', inplace=True)

# Concatenate 'business' and 'month' columns to create x-axis labels
trips_count_df['business_month'] = trips_count_df['business'] + '-' + trips_count_df['month'].astype(str)

# Plot the histogram
plt.figure(figsize=(10, 6))
bars = plt.bar(trips_count_df['business_month'], trips_count_df['count'])
plt.xlabel('Business-Month')
plt.ylabel('Trip Count')
plt.title('Business-Month vs Trip Count')
plt.xticks(range(len(trips_count_df)), trips_count_df['business_month'], rotation=45)
# Set y-axis scale to linear
plt.yscale('log')

# Add count values above each bar
for bar, count in zip(bars, trips_count_df['count']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(count),
             ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

### 2.2 Calculating Platformâ€™s Profits for each Month

```python
# sort values by month
platform_profits_df.sort_values(by='month', inplace=True)

# Concatenate 'business' and 'month' columns to create x-axis labels
platform_profits_df['business_month'] = platform_profits_df['business'] + '-' + platform_profits_df['month'].astype(str)

# Plot the histogram
plt.figure(figsize=(10, 6))
bars = plt.bar(platform_profits_df['business_month'], platform_profits_df['sum(rideshare_profit)'])
plt.xlabel('Business-Month')
plt.ylabel('Profits')
plt.title("Business-Month vs Platform's Profits")
plt.ticklabel_format(style='plain', axis='y')  # Display y-axis tick labels in plain format

# Add count values above each bar
for bar, count in zip(bars, platform_profits_df['sum(rideshare_profit)']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), "{:.2f}".format(count),
             ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

### 2.3 Calculating Driverâ€™s Earnings for each Month

```python
# sort values by month
driver_earnings_df.sort_values(by='month', inplace=True)

# Concatenate 'business' and 'month' columns to create x-axis labels
driver_earnings_df['business_month'] = driver_earnings_df['business'] + '-' + driver_earnings_df['month'].astype(str)

# Plot the histogram
plt.figure(figsize=(10, 6))
bars = plt.bar(driver_earnings_df['business_month'], driver_earnings_df['sum(driver_total_pay)'])
plt.xlabel('Business-Month')
plt.ylabel('Profits')
plt.title("Business-Month vs Driverâ€™s Earnings")
plt.ticklabel_format(style='plain', axis='y')  # Display y-axis tick labels in plain format
plt.yscale('log')
# Add count values above each bar
for bar, count in zip(bars, driver_earnings_df['sum(driver_total_pay)']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), "{:.2f}".format(count),
             ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```
